## Working with pipelines
The most basic object in 🤗 Transformers library. Connects model with its necessary pre- and postprocessing steps, allowing to directly input any text and get an answer.

```python
# one sentence
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
classifier("I do not feel good right now...")
```

```python
# multiple sentences
classifier(
    ["I've been waiting for a HuggingFace course my whole life.",
     "I hate this so much!",
     "What have you done?",
     "Do I look silly in this?"]
)
```

### Zero-shot classification
It's called `zero-shot` because you do not need your model to be fine-tuned on the data you want to use. It can directly return probability scores of any labels.

This pipeline allows you to specify which labels to use for the classification so there's no need to rely on the labels of the pretrained model.

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification")
classifier(
    "This course is not about Transformers",
    candidate_labels=['tests', 'learning', 'medicine', 'sports']
)
```

### Text generation
In this `pipeline` you provide a prompt and the model will automatically complete it by generating the remainder.

To control how many different sequences are generated use `num_return_sequences`. The total length of a sentence is limited by `max_length`.

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model='distilgpt2'
)
generator(
    "In this course, I'll let you know of",
    max_length=25,
    num_return_sequences=3
)
```

### Mask filling
The main idea behind this task is to fill in the blanks of a given text. The `top_k` controls the number of how many possibilities will be shown.

**Note:** It always good to verify the `mask token` for a model you'll be using as it's very likely for different models to use different masks.

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker(
    "This course will teach you about <mask> models.",
    top_k=3
)
```

### Named entity recognition (NER)
This is a task where model has to find parts of text that correspond to enitites like persons, organizations or places. In abbreviated form it's usually addressed as **NER**.

The `grouped_entities` tells the pipeline to regroup together parts of sentence that correspond to the same entity.

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    grouped_entities=True)

ner("My name is Paweł, I live in Wrocław and work at Endress+Hauser")
```

### Question answering
This kind of pipeline answers questions using information from a given context.

**Note:** This pipeling works by EXTRACTING information from te provided content.

```python
from transformers import pipeline

qa = pipeline('question-answering')
qa(
    question="Where do you work?",
    context="I like to work in my garden, but my main job is Sales Manager"
)
```

### Summarization
This task is focused on reducing a text into a shorter text while keeping the important aspects of it.

Similarly as in *text generation*, it's possible to specify `max_length` or a `min_length` of the result.

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other
    industrial countries in Europe and Asia, continue to encourage and advance
    the teaching of engineering. Both China and India, respectively, graduate
    six and eight times as many traditional engineers as does the United States.
    Other industrial countries at minimum maintain their output, while America
    suffers an increasingly serious decline in the number of engineering graduates
    and a lack of well-educated engineers.
""",
    max_length=80,
    min_length=70
)
```
